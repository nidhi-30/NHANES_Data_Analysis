```
from keras.utils.np_utils import to_categorical
from keras.optimizers import RMSprop,Adam
```

```
# encode class values as integers
encoder = LabelEncoder()
encoder.fit(hypertensionData['Target'])
encoded_Y = encoder.transform(hypertensionData['Target'])
# convert integers to dummy variables (i.e. one hot encoded)
y = to_categorical(encoded_Y)
```

```
hypertensionX_train, hypertensionX_test, dataY_train, dataY_test = train_test_split(hypertensionData.drop('Target', axis=1), y, 
                                                                                    test_size=0.33, random_state=42)
```

```
hypertensionX_train = hypertensionX_train.to_numpy()
hypertensionX_test = hypertensionX_test.to_numpy()
```

```
dataX_train = np.concatenate((extracted_model_demoX_train, extracted_model_dietX_train, extracted_model_examX_train, 
                              extracted_model_labX_train, extracted_model_questionX_train, hypertensionX_train),
                             axis = 1)
```

```
dataX_train.shape
```

```
dataX_test = np.concatenate((extracted_model_demoX_test, extracted_model_dietX_test, extracted_model_examX_test, 
                             extracted_model_labX_test, extracted_model_questionX_test, hypertensionX_test),
                             axis = 1)
```

```
dataX_test.shape
```

```
model = Sequential()
model.add(Conv1D(256, 5, activation='relu', input_shape=(817, 1)))
model.add(MaxPooling1D(2))
model.add(Conv1D(128, 3, activation='relu'))
model.add(MaxPooling1D(2))
model.add(Conv1D(64, 3, activation='relu'))    
model.add(MaxPooling1D(2))
model.add(Flatten())
model.add(Dense(128, activation='relu'))
model.add(Dense(64, activation='relu'))
model.add(Dense(32, activation='relu'))
model.add(Dense(16, activation='relu'))
model.add(Dense(2, activation='softmax'))
print(model.summary())
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```
dataX_train = dataX_train.reshape(dataX_train.shape[0], dataX_train.shape[1], 1)
dataX_test = dataX_test.reshape(dataX_test.shape[0], dataX_test.shape[1], 1)
```

```
y_pred = model.predict(dataX_test)
#Converting predictions to label
pred = list()
for i in range(len(y_pred)):
    pred.append(np.argmax(y_pred[i]))
#Converting one hot encoded test label to label
test = list()
for i in range(len(dataY_test)):
    test.append(np.argmax(dataY_test[i]))
```
